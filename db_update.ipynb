```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('food2.csv', encoding='cp949')
df
```

```python
kcal = df['kcal'] == 'TRACE'
carbo = df['carbo'] =='TRACE'
protein = df['protein'] =='TRACE'
fat = df['fat'] =='TRACE'

find = df[kcal|carbo|protein|fat].index
df = df.drop(find)
```

```python
df['id'] = [x for x in range(1,len(df)+1)]
```

```python
df = df[['id', 'name', 'serving_wt', 'kcal', 'carbo', 'protein','fat', 'company']]
df
```

```python
df.to_csv('food3.csv', encoding='cp949', index=False)
```

```python
conn = sqlite3.connect("db.sqlite3")
```

```python
cur = conn.cursor()
```

```python
check = df.values.tolist()
check
```

```python
cur.executemany( 'INSERT INTO FoodInfo_table VALUES (?, ?, ?, ?, ?, ?, ?, ?)', 
                check )
```

```python
conn.commit()
conn.close()
```
